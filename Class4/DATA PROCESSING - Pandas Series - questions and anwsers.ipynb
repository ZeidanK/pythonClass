Load the file "/content/tweeter username and followers.csv" from Moodle and read the content in a Pandas series named "username_and_followers". Print how many usernames and how many followers are in total in the "username_and followers" series.

#Problem 2
Build a tweet_users_dict dictionary with two lists with keys: Username and Followers. Create two lists of Usernames and Followers from the list you read in previous question.
Note: you must skip the first read element with description of columns


```python
# Define an dictionary with two keys Username and Followers, each one with two empty lists.
tweet_users_dict = { 'Username':list(username_and_followers ['user_screen_name' ].values), 'Followers': list(username_and_followers ['user_followers_count' ].values)}
```

```python
tweet_users_dict
```

# Problem 3
 Build  list with lists having the index number, the Username and the number of followers of all Followers including the words 'love' or 'hate' (Note: those words can be written with any combination of lowercases and uppercases)


```python
# Define an empty list that will contain the list of lists of filtered elements (including love or hate)
output_list = []

# Go over all Usernames
for idx in range(len(tweet_users_dict['Username'])):
  # if the Username converted to lowercase includes love or hate add it to the list
  if 'love' in tweet_users_dict['Username'][idx].lower() \
  or 'hate' in tweet_users_dict['Username'][idx].lower() :
    print(idx,tweet_users_dict['Username'][idx],tweet_users_dict['Followers'][idx])
    # Add to the list a new list compound of three elements: Username and Followers
    output_list.append([tweet_users_dict['Username'][idx],tweet_users_dict['Followers'][idx]])
```

```python
output_list
```

# Problem 4
Write the content of the list of elemetns of previous question (Problem 3) into a file name "output_file.csv" with every element in a different line and each part of the element separated by commas (,)

```python
# Define an output file
output_df = pd.DataFrame(output_list,columns=["username","followers"])
output_df.to_csv("output_file.csv")
```
