## An example for using pandas: diamonda data set
## **YOU ARE NOT SUPPOSED TO UNDERSTAND THE CODE, IT IS JUST AN EXAMPLE**

```python
import pandas as pd  # this is how usually we will import pandas
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

df = sns.load_dataset('diamonds')
df.head(10)
```

##Investigate data

```python
plt.figure(figsize=(10,10))
plt.xlabel("price", fontsize=15)
plt.ylabel("count", fontsize=15)
df['price'].hist()
plt.show()
```

##Choose rows by col value

```python
df[df['cut'] == 'Ideal'].head()
```

##Sort values by specific col

```python
df[df['cut'] == 'Ideal'].head().sort_values(by=['price'])
```

##Check assumpitons - diamond color affects the price

```python
plt.figure(figsize=(15,15))
df[['color', 'price']].groupby(['color']).mean().plot.bar(y="price")
plt.show()
```

#Pandas series
* ### One dimensional array like object containing an array of data and an associated array of data labels, called index
* ### A mix between regular python lists and dictionaries

```python
import pandas as pd  # the common way to import pandas
import numpy as np

s = pd.Series([4, 7, -5, 3])
print(s)
```

```python
s = pd.Series(['kuku', 7, 'blabla', 3])
print(s)
```

##Indexing and slicing - similar to python arrays

```python
print(s[0])
```

```python

print(s[0:2])
```

###Choose by index list

```python
print(s[[1,3]])
```

### We can get the values and the index as lists (similar to dicts)

```python
print(s.values)
print(list(s.index))
```

##Index can be any type we want

```python
s = pd.Series([4, 7, -5, 5], index=[5, 'b', 5.5, True])
print(s)
```

```python
print(s.values)
print(list(s.index))
print(s['b'])
print(s[True])
```

### Series index can be altered in place by assignment (but must be of the same length)


```python
s.index = ['Bob', 'Steve', 'Jeff', 'Ryan']
print(s)
```

###Mathematical operations can be done on a series similar to np arrays

```python
print(s*4)
```

```python
print(s+s)
```

```python
print(1/s)
```

```python
print(np.exp(s))
```

###Boolean indexing

```python
print(s)
```

```python
print(s>0)
```

```python
print(s[s>0])
```

### We can specify which values from dict we want to take


```python
sdata = {'Ohio': 35000, 'Texas': 71000, 'Oregon': 16000, 'Utah': 5000}
states = ['California', 'Ohio', 'Oregon', 'Texas']
s4 = pd.Series(sdata, index=states)
print(s4)
```

```python
print(s4.isnull())
```

```python
print(s4[s4.isnull()])
```

```python
print(s4[s4.notnull()])
```

```python
print(s4[~s4.isnull()])
```

```python
print(s4.isnull()) # pd automatically fills NaN when he can't find a value
print(s4[s4.isnull()])
print(s4[s4.notnull()])
```

###Series object and its index have name attribure, which intergrates with other key areas of pandas functionality. Meaning that class series has an attribute name.

```python
s4.name = 'population'
s4.index.name = 'state'
print(s4)
```
