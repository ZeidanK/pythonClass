## Arguments
* ### Information can be passed into functions as arguments
* ### We add them inside the parentheses after function name


```python
def func_with_arg(x, y, z):
  a = x+y+z
  return a

print(func_with_arg(1,2,3))
print(func_with_arg(4,5,6))
print(func_with_arg("4","5","6"))
print(func_with_arg(3,2,1))
print(func_with_arg("aaa","bbb","ccc"))
```

```python
print(func_with_arg(3.34,2,1.11))
```

# Function example:
Write code that receives as input three numbers A, B and C and prints the factorial of A (A!) , the factorial of B (B!) and the factorial of C (C!).

n! = 1 x 2 x 3 x .... x (n-1) x n

```python
def fact(n):
    n_fact = 1
    if n < 0:
      print("Error")
    elif n == 1:
        print(1)
    else:
      for i in range(2,n+1):
        n_fact*=i
      print(n_fact)
```

```python
fact(9)
```

```python
fact(1)
```

```python
fact(-5)
```

```python
x = fact(9)
```

```python
print(x)
```

```python
x = fact(5)
```

```python
print(x)
```

# Function example:
Write code that receives as input three numbers A, B and C and returns the factorial of A (A!) , the factorial of B (B!) and the factorial of C (C!).

n! = 1 x 2 x 3 x .... x (n-1) x n

```python
def fact(n):
    n_fact = 1
    if n < 0:
      return "ERROR"
    elif n == 1:
        return 1
    else:
      for i in range(2,n+1):
        n_fact*=i
      return n_fact
```

## Function example:
### Write a function that recives a string of a sentence and returns a new string of the sentence reversed.
### **input**: one two three four
### **output**: four three two one

```python
def reverse_sentece(sentence):
  split_sentence = sentence.split(" ")
  print(split_sentence)
  reversed_lst = split_sentence[::-1]
  print(reversed_lst)
  rev_sentence = " ".join(reversed_lst)
  return rev_sentence

sentence = "one two three four"
print(reverse_sentece(sentence))
```

## Another example:
### Write a function that recives a string and prints a histogram of the letters appearance (the number of times each letter appears).
### **input**: "this is a string"
### **will print**:
###t    2
###h    1
###i    3
###s    3
###$\space$     3
###a    1
###r    1
###n    1
###g    1


```python
def letters_hist(string):
  letters_dict = {}
  for letter in string:
    if letter in letters_dict.keys():
      letters_dict[letter] += 1
    else:
      letters_dict[letter] = 1
  for key in letters_dict.keys():
    print(key, "  ", letters_dict[key])

sentence = "this is a wonderful day, i would like to take a walk outside"
letters_hist(sentence)
```

# Another Example
Write a function print_list that receives as parameter a list LIST, the starting position START and a step STEP and prints the elements of LIST starting at position START with increment STEP

```python
def print_list(LIST, START = 1, STEP = 1):
  if START < 0 or START >= len(LIST):
    return "ERROR"
  for index in range(START-1,len(LIST),STEP):
    print(LIST[index])
```

```python
print_list(["This","is",5,"an","888","example",987])
```

```python
print_list(["This","is",5,"an","888","example",987],2,2)
```
