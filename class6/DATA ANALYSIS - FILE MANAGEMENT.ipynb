# Question 1
## How to create a file based on "members", replacing city_id by city name?


```python
# Merge dataframes based on the specified columns
members_full_df = pd.merge(members_df, cities_df, left_on='city_id', right_on='id')
```

```python
members_full_df.info()
```

```python
members_full_df.head()
```

```python
# remove columns id_y and city_id
members_full_df = members_full_df.drop(["id_y","city_id"],axis=1)
```

```python
members_full_df.head()
```

```python
# rename column id_x to id
members_full_df = members_full_df.rename({"id_x":"member_id","city":"member_city"},axis=1)
```

```python
members_full_df.head()
```

```python
members_full_df [ members_full_df [ 'member_id'] == 25]
```

```python
members_full_df = members_full_df.set_index("member_id")
```

# Question 2
## Complete Restaurants file merging with restaurant_types and cities

```python
restaurants_df = pd.read_csv(file_path+'restaurants.csv')
```

```python
restaurants_df.info()
```

```python
restaurants_df.head()
```

```python
# Merge dataframes based on the specified columns
restaurants_full_df = pd.merge(restaurants_df, cities_df, left_on='city_id', right_on='id')
```

```python
restaurants_full_df.head()
```

```python
restaurants_full_df = restaurants_full_df.drop(["id_y","city_id"],axis=1)
```

```python
restaurants_full_df = restaurants_full_df.rename({"id_x":"restaurant_id","city":"restaurant_city"},axis=1)
```

```python
restaurants_full_df.head()
```

```python
restaurant_types_df = pd.read_csv(file_path+'restaurant_types.csv')
```

```python
restaurant_types_df.info()
```

```python
restaurant_types_df.head()
```

```python
restaurants_full_df = pd.merge(restaurants_full_df, restaurant_types_df, left_on='restaurant_type_id', right_on='id')
```

```python
restaurants_full_df.head()
```

```python
restaurants_full_df = restaurants_full_df.drop(["restaurant_type_id"],axis=1)
```

```python
restaurants_full_df
```

## Complete Orders file merging with members and restaurants


```python
orders_df = pd.read_csv(file_path+'orders.csv')
```

```python
orders_df.info()
```

```python
orders_df.head()
```

```python
orders_full_df = pd.merge(orders_df,  members_full_df, left_on='member_id', right_on='member_id').sort_values(by="id")
```

```python
orders_full_df.head()
```

```python
orders_full_df = pd.merge(orders_full_df,  restaurants_full_df, left_on='restaurant_id', right_on='restaurant_id').sort_values(by="id_x")
```

```python
orders_full_df.head()
```

```python
orders_full_df = orders_full_df.drop("id_y",axis=1)
```

```python
orders_full_df = orders_full_df.rename({"id_x":"order_id"},axis=1)
```

```python
orders_full_df.head()
```
