# Regular Expression module (re) examples

Finding Words:

pattern = r'\bS\w*n\b' defines a pattern to find words that start with 'S' and end with 'n'. \b denotes a word boundary, \w* matches any number of word characters.
re.findall(pattern, text) finds all occurrences matching the pattern in the text.

```python
import re

# Example text
text = "The rain in Spain falls mainly in the plain. Shain"

# 1. Find all occurrences of a word that starts with 'S' and ends with 'n'
pattern = r'\bS\w*n\b' #understand this pattern in the text
matches = re.findall(pattern, text)
print("Words starting with 'S' and ending with 'n':", matches)
```

Replacing Text:

pattern = r'rain' defines a pattern to find the word 'rain'.
re.sub(pattern, 'snow', text) replaces all occurrences of 'rain' with 'snow'.

```python
# 2. Replace all occurrences of 'rain' with 'snow'
pattern = r'rain'
new_text = re.sub(pattern, 'snow', text)
print("Text after replacement:", new_text)
```

Extracting Words Starting with 'i', 't', or 'r':

pattern = r'\b[iItrR]\w*\b' defines a pattern to match words starting with 'i', 't', or 'r'. \b denotes a word boundary, [iItrR] matches any of the specified letters, and \w* matches any number of word characters.
re.findall(pattern, text, re.IGNORECASE) extracts all words from the text that start with 'i', 't', or 'r', ignoring case.

```python
# 3. Extract words starting with 'i', 't', or 'r' (case-insensitive)
pattern = r'\b[itr]\w*\b'
words = re.findall(pattern, text, re.IGNORECASE)
print("Words starting with 'i', 't', or 'r':", words)
```

Matching Text Start:

pattern = r'^The' defines a pattern to check if the text starts with 'The'. ^ denotes the start of the string.
re.match(pattern, text) checks if the text matches the pattern from the beginning.
This update focuses on extracting words starting with specified letters and makes the search case-insensitive.

```python
# 4. Check if the text starts with 'The'
pattern = r'^The'
match = re.match(pattern, text)
if match:
    print("The text starts with 'The'")
else:
    print("The text does not start with 'The'")

print(match)
print(bool(match))  
```

# Function defined to clean text from URLs and special characters

```python
# define a function that removes URLs and special characters from a text
import re
def remove_urls_and_special_characters(txt):
    """Replace URLs and special characters found in a text string with nothing;
       Replace multiple spaces between words by a single space

    Parameters
    ----------
    txt : string
        A text string that you want to parse and remove urls.

    Returns
    -------
    The same txt string with url's removed.
    """

    return " ".join(re.sub("([^0-9A-Za-z \t])|(\w+:\/\/\S+)", "", txt).split())

    #return re.sub("([^0-9A-Za-z \t])|(\w+:\/\/\S+)", "", txt) what is the difference from previous line?
```

```python
print(tweets_df [ 'Text'].values)
```

```python
print([ remove_urls_and_special_characters(text) for text in tweets_df [ 'Text']])
```

```python
tweets_df [ 'Clean_text'] = [ remove_urls_and_special_characters(text) for text in tweets_df [ 'Text']]
```

# Task 2

Translate all non-English texts to English using provided function translate_to_English creating a new column "Translated_text". Save the new table under the name @Isaac_Herzog_user_tweets_CLEAN_TRANSLATED.xlsx

```python
!pip install deep_translator
```

```python
from deep_translator import GoogleTranslator
# Define a function that translate any text to English
def translate_to_English (txt):
    try:
      translated_txt = GoogleTranslator(source='auto', target='en').translate(txt)
    except:
      translated_txt = "TRANSLATION ERROR"
    return translated_txt
```

```python
tweets_df ['Translated_text'] = [ translate_to_English(text) for text in tweets_df [ 'Clean_text']]
```
