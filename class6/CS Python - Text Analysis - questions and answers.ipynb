# Question 3

First, create a list having all words of all tweets.
Then, create a filtered list 'list_of_filtered_words' having only words not stop-words and not starting nor @ neither ^.
Use an auxiliar function create by you named filter_word receiving a word as input and returing True is the word is the stop-wod or start with @ neither ^.

```python
import nltk
#Read stop words from nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = list(set(stopwords.words('english')))

# View a few words from the set
print(stop_words[0:20])
```

```python
def filter_word(text,stopwords = stop_words):
  if text[0] == "@" or text[0] == "^":
    return True
  if text.lower() in stop_words:
    return True
  return False
```

```python
list_of_all_words = [item for sublist in df_copy [ 'list_of_words' ] for item in sublist]
```

```python
list_of_all_filtered_words = [ word  for word in list_of_all_words if filter_word(word) == False]
```

```python
list_of_all_filtered_words
```

# Question 4
Display Top 20 frequent non-stop words words

```python
list_of_all_filtered_words_df = pd.DataFrame(data=list_of_all_filtered_words,columns=['Word'])
```

```python
list_of_all_filtered_words_df
```

```python
# Count the occurrences of each word
word_counts = list_of_all_filtered_words_df['Word'].value_counts()

# Convert to DataFrame for easier plotting with seaborn
word_counts_df.columns = ['Word', 'Count']
```

```python
# Plot graph of Top 20 mentions
import plotly.express as px
fig = px.bar(data_frame = word_counts_df[:20],   # take the frequency of the first 20 frequent words
             y="Word",                        # take the first 20 frequent words
             x="Count",                        # x axis = Number of words
             labels={
                    "y": "Word",                         # Set index (y) axis name to words
                    "count": "Frequency",               # Set index (x) axis name to # of words
                 },
             orientation='h',                         # layout vertical
             color="Count",                         # set color scheme according to words
             text="Count")                          # Display the # of words as labels of the bar
fig.update_layout(width=1000,                          # set width of graph
                  height=600,                         # set height of graph
                  title="Top 20 metioned words",            # set title
                  yaxis=dict(autorange="reversed"))   # reverse order
fig.show() # show the graph
```
