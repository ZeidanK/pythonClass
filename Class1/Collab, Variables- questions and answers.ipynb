#Problem 14
Write a Python program to accept a filename from the user and print the extension of that file
Example: Input: Image.jpg -> Output:jpg
Assumptiom: file name has an unique "." character

```python
my_filename = input("Enter a filename:")
```

```python
point_location = my_filename.find(".")
```

```python
file_extension = my_filename [ point_location + 1 :]
```

```python
print("File extension:",file_extension)
```
