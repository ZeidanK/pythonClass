#Problem 4
Input a string a replace all blank spaces by "_" (using split and join). Input:"Enter a string:" -> Output: This_is_a_sentence

```python
my_string = input("Enter a sentence: ")
```

```python
my_string_after_split = my_string.split()
```

```python
print(my_string_after_split)
```

```python
print("_".join(my_string_after_split))
```

```python
my_string = input("Enter a sentence: ")
```

```python
output_string = ""
for w in my_string:
  if w == " ":
    print("_")
    output_string = output_string + "_"
  else:
    print(w)
    output_string = output_string + w

  print("Until now output_string = ",output_string)
```

```python
print(output_string)
```
