# Python Lists (arrays):
* ## Necessary when we use many variables and we want to save them together
* ## A sequence of values of any type
* ## In one list we can have multiple types
* ## The simplest way to create a list is to enclose elements in square brackets ([ ])
* ## Don't have a predefined size

```python
my_first_list = [1, 2.0, "some string", True]
print(type(my_first_list))
print(my_first_list)

# a list within another list is called nested
my_first_nested_list = [1, 2.0, 'some string', True,[1 , False]]
print(my_first_nested_list)
```

```python
my_first_nested_list[4]
```

##List operations

```python
my_first_nested_list = [1, 2.0, "some string", True, [1, False]]
my_first_nested_list += [1, 2, 3]
print(my_first_nested_list)
print([0] * 4)
print([1, 2, 4] * 4)
```

#While loop
* ## Used When we need to do identical or similar operations many times
* ## A while loop has a boolean condition - while <condition\>. The loop will repeat itself until the condition is False

```python
i = 0
while i < 10:
  print(i)
  i += 1
```

##While example - find the first number n that the result of $n^3$ is larger than 10,000

```python
n = 1
while n**3 < 10000:
  print("******",n)
  n += 1
print("the number", n, "in the power of 3 is larger than 10000, n^3 = ", n**3)
```

### For loop example - get a number from user, check it is really a number (only digits) and check if it is a prime number

```python
N = ""
# make sure we get a number
while type(N) != int:
  N = input("enter a number")
  if N.isdigit():
    N = int(N)
  else:
    print("enter only numbers")

is_prime = True
# check if it is a prime number
for i in range(2, N):
  print(i)
  if N % i == 0:
    is_prime = False
    break

if is_prime:
  print(N, "is prime!")
else:
  print(N, "is not prime!")
```

#Break and continue

##Break:
* ### Terminates the current loop and resumes execution at the next statement
* ### The most common use for break is when some external condition is triggered requiring a hasty exit from a loop
* ### The break statement can be used in both while and for loops

```python
for i in range(10):
  if i == 7:
    break
  print(i,end=" ")
```

##Let's check if a number is prime:
### we can stop when we find a number that his mod is 0

```python
num_str=input("Enter a positive integer: ")
num = int(num_str)
is_prime = True

for i in range(2, int(num**0.5+1)):
  if (num%i == 0):
    is_prime = False
    break

if is_prime:
  print("a prime!")
else:
  print("not a prime")
```

##Continue:
* ### Stops the current iteration and "jumps" to the next
* ### 7 "BOOM"

```python
for i in range(1,101):
  if i%7 != 0 and i%10 != 7:
    continue
  print(i, " BOOM!!!")
```

```python
for i in range(1,101):
  if i%7 == 0 or i%10 == 7:
    print(i, " BOOM!!!")
```

#Nested loops:
* ## A loop inside a loop body

##Print a rectangle of *
* ## Recive num of rows and cols from user

```python
print("We'll draw a rectangle")
row_len=int(input("Enter row length: "))
num_rows=int(input("Enter number of rows: "))
for i in range(num_rows):
  for j in range(row_len):
    print('*', end='')
  print()
```

## Another Example:
### Print all prime numbers in a given range

```python
print("This program finds all primes in a range")
lower_str = input("Enter lower bound: ")
lower = int(lower_str)
upper_str = input("Enter upper bound: ")
upper = int(upper_str)
count=0

for num in range(lower, upper+1):
  is_prime = True
  for i in range(2, int(num**0.5+1)):
    if (num%i == 0):
      is_prime = False
      break

  if not is_prime: # same as if is_prime == False:
    continue
  count+=1

  if count%10==0:
    print(num)
  else:
    print(num,end=' ')
```
