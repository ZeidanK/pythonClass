#Problem 2
You have two lists of same length: keys and data. Build a program that gets keys and data as parameters and returns a dictionary when entry has a key in keys list and data in data list.

Example

keys = [ 12345, 23454, 32345, 98709, 45365 ]

data = [ ['Maths', 90], ['English','90'], ['Maths',87], ['Sport',60], ['Hebrew',78] ]

Output: {12345: ['Maths', 90], 23454: ['English', '90'], 32345: ['Maths', 87], 98709: ['Sport', 60], 45365: ['Hebrew', 78]}

```python
keys = [ 12345, 23454, 32345, 98709, 45365 ]

data = [ ['Maths', 90], ['English','90'], ['Maths',87], ['Sport',60], ['Hebrew',78] ]
```

```python
d = {}
for i in range(len(keys)):
  #print(keys[i],data[i])
  d [ keys[i] ] = data[i]
print(d)
```

```python
d = { keys[i]:data[i] for i in range(len(keys))}
print(d)
```
