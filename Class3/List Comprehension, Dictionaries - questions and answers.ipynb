#Problem 2
You have two lists of same length: keys and data. Build a function that gets keys and data as parameters and returns a dictionary when entry has a key in keys list and data in data list.

Example

keys = [ 12345, 23454, 32345, 98709, 45365 ]

data = [ ['Maths', 90], ['English','90'], ['Maths',87], ['Sport',60], ['Hebrew',78] ]

Output: {12345: ['Maths', 90], 23454: ['English', '90'], 32345: ['Maths', 87], 98709: ['Sport', 60], 45365: ['Hebrew', 78]}

```python
keys = [ 12345, 23454, 23454, 12345, 45365 ]
data = [ ['Maths', 90], ['English','90'], ['Maths',87], ['Sport',60], ['Hebrew',78] ]
```

```python
my_dict = {}
for idx in range(len(keys)):
  if keys[idx] in my_dict.keys():
    my_dict [ keys[idx] ].append(data [ idx ])
  else:
    my_dict [ keys[idx] ] = [ data [ idx ] ]
```

```python
print(my_dict)
```

#Problem 3
Write a function that gets a dictionary with the format of te output of previous question (Problem 2) and returns a list with the average of all grades of each student.

```python
my_dict = {12345: [['Maths', 90], ['Sport', 60]], 23454: [['English', '90'], ['Maths', 87]], 45365: [['Hebrew', 78]]}
```

```python
output_list = []
```

```python
for single_key in my_dict.keys():
  grade_list = my_dict [ single_key ]
  grades = ([int(record[1]) for record in grade_list])
  average = float(sum(grades)) / len(grades)
  output_list.append([single_key,average])
```

```python
print(output_list)
```

#Problem 4

Use list comprehension

Output: [1, 0, 15, 0, 45, 0, 91, 0, 153, 0, 231, 0, 325, 0, 435, 0, 561, 0, 703, 0, 861, 0, 1035, 0, 1225]
