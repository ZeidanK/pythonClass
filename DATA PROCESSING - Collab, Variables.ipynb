### Math operations

```python
# Math operations

print("5+5.5 =", 5+5.5)
print("5*5 =", 5*5)
print("5**2 =", 5**2)
print("10 % 3 =", 10 % 3)
print("100**0.5 =", 100**0.5)##sqrt of 100 
print("10/3 =", 10/3)
print("type(10/3) =", type(10/3))
print("10//3 =", 10//3)
print("type(10//3) =", type(10//3))
```
